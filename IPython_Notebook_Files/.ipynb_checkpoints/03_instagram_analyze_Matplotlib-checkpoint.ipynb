```python
import requests
import requests
from pandas.io.json import json_normalize
import pandas as pd
import datetime



def instagram_data(query):
    base_url = "https://api.instagram.com/v1"
    CLIENT_ID = '768fcf1f36c94eb08506bae0a9caffa3'

    url = '{0}/tags/{1}/media/recent?client_id={2}&count=30'.format(
        base_url, query, CLIENT_ID)

    r = requests.get(url)
    j = r.json()
    
    results = []
    if 'data' in j: 
        data = j['data']
        df_instance = json_normalize(data)
        results.append(df_instance)
        
    df = pd.DataFrame().append(results)

    def instagram_data_clean():
        df = df
        df['created_time'] = [
        y.replace(y, datetime.datetime.fromtimestamp(int(str(y))).strftime(
        '%Y-%m-%d %H:%M:%S')) for y in df['created_time']]

        cols = [
        'user.username',
        'caption.text',
        'tags',
        'comments.count',
        'likes.count',
        'filter',
        'type',
        'created_time',
        'user.full_name',
        'user.id',
        'link',
        'location.latitude',
        'location.longitude'
    ]

        df_cols = df[cols]
        df_2 = df_cols.rename(columns=lambda x: x.replace('.',' ').title())

        return df_2
```

```python
instagram_data('nyc')
```

```python
df_2 = instagram_data_clean(instagram_data('nyc'))
```

```python
# New imports to work with
import matplotlib.pyplot as plt

%matplotlib inline
```

You may have notice our functions building. Now we could pull data from Instagram, select the fields we want to work with, clean the fields and output graphs. See one graph below using matplotlib.

```python
df_2['Likes Count'].plot(kind='bar', alpha=.55)
plt.title("Total Like Count Per Post", fontsize=20)
plt.xlabel('Most Recent to Least Recent')
plt.ylabel('Total Likes')
```
